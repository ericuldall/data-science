```python
broken_emails = pd.read_csv('/home/eric/emails.csv')
origin_files = ['ccjrcs-l5m-5-19-2015-all.csv', 'ccvisitors-l5m---rm-vq-6-10-15-all.csv', 'ccvisitors-l5m-vq-6-3-15-all.csv', 'ccvisitors-l5m-vq-june-2015-all.csv', 'cpus-jrcs-21-6-16-15-all.csv', 'ccjrcs-l5m-5-27-2015-all.csv', 'ccvisitors-l5m-vq-6-15-15-all.csv', 'ccvisitors-l5m-vq-april-2015-all.csv', 'ccvisitors-l5m-vq-may-2015-all.csv', 'ccvisitors-l5m-6-11-15-all.csv', 'ccvisitors-l5m-vq-6-2-15-all.csv', 'ccvisitors-l5m-vq---june-2015-all.csv', 'cpus-clickers-may-(l5m-)-all.csv']
origin_datasets = []
for file in origin_files:
    ds = pd.read_csv('/home/eric/Sendy_FYC/'+file)
    ds.columns = ['FIRST', 'EMAIL', 'LAST', 'CITY', 'STATE', 'ZIP', 'KEYWORD', 'PARTNER']
    origin_datasets.append(ds)
```

```python
final = pd.concat(origin_datasets)
```

```python
final
```

```python
broken_emails.columns = ['EMAIL']
broken_emails
```

```python
fixed = pd.merge(final, broken_emails, how='inner', on=['EMAIL']).drop_duplicates(['EMAIL'])
fixed
```

```python
fixed = fixed.fillna('')
for row in fixed.iterrows():
    print('UPDATE auth_users SET first_name = "'+str(row[1]['FIRST'])+'", last_name = "'+str(row[1]['LAST'])+'", city = "'+str(row[1]['CITY'])+'", state = "'+str(row[1]['STATE'])+'", zip = "'+str(row[1]['ZIP'])+'", pid = "'+str(row[1]['PARTNER'])+'" WHERE email = "'+str(row[1]['EMAIL'])+'";')
```
